```python
for code in code_list:
    loaded_data = pd.read_csv('./closed/'+ code + ".csv", sep='\t', header=None)
    loaded_data_length = len(loaded_data[1])
    q1_2015 = loaded_data[1][0:round(loaded_data_length/12)].mean()
    q2_2015 = loaded_data[1][round(loaded_data_length/12)+1:round(loaded_data_length/12)*2].mean()
    q3_2015 = loaded_data[1][round(loaded_data_length/12)*2+1:round(loaded_data_length/12)*3].mean()
    q4_2015 = loaded_data[1][round(loaded_data_length/12)*3+1:round(loaded_data_length/12)*4].mean()
    q1_2016 = loaded_data[1][round(loaded_data_length/12)*4+1:round(loaded_data_length/12)*5].mean()
    q2_2016 = loaded_data[1][round(loaded_data_length/12)*5+1:round(loaded_data_length/12)*6].mean()
    q3_2016 = loaded_data[1][round(loaded_data_length/12)*6+1:round(loaded_data_length/12)*7].mean()
    q4_2016 = loaded_data[1][round(loaded_data_length/12)*7+1:round(loaded_data_length/12)*8].mean()
    q1_2017 = loaded_data[1][round(loaded_data_length/12)*8+1:round(loaded_data_length/12)*9].mean()
    q2_2017 = loaded_data[1][round(loaded_data_length/12)*9+1:round(loaded_data_length/12)*10].mean()
    q3_2017 = loaded_data[1][round(loaded_data_length/12)*10+1:round(loaded_data_length/12)*11].mean()
    q4_2017 = loaded_data[1][round(loaded_data_length/12)*11+1:round(loaded_data_length/12)*12].mean()
    
    print(code, ',q1_2015,', round(q1_2015,2))
    print(code, ',q2_2015,', round(q2_2015,2))
    print(code, ',q3_2015,', round(q3_2015,2))
    print(code, ',q4_2015,', round(q4_2015,2))
    print(code, ',q1_2016,', round(q1_2016,2))
    print(code, ',q2_2016,', round(q2_2016,2))
    print(code, ',q3_2016,', round(q3_2016,2))
    print(code, ',q4_2016,', round(q4_2016,2))
    print(code, ',q1_2017,', round(q1_2017,2))
    print(code, ',q2_2017,', round(q2_2017,2))
    print(code, ',q3_2017,', round(q3_2017,2))
    print(code, ',q4_2017,', round(q4_2017,2))
```
