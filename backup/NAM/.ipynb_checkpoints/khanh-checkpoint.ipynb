```python
def write_csv(name, data, basePath):
    with open(basePath + name + ".csv", "w", newline='', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file, delimiter=',')
        for line in data:
            writer.writerow(line)
```

```python
def get_code_divided(category, url='http://www.cophieu68.vn/categorylist_detail.php?category=^'):
    code_list_divided = []
    result = requests.request("GET", url , params={"category": "^"+ category})
    soup = BeautifulSoup(result.content, "html5lib")
    for data in soup.find_all("strong"):
        code_list_divided.append(data.get_text())
    return code_list_divided
```

```python
types_report = ["bs", "ist", "cf"]
```

```python
for cat_index, category in enumerate(category_list):
    bs = []
    ist = []
    cf = []
    for type_report in types_report:
        for code in get_code_divided(category):
            params = {
                    "id": code,
                   "lang": "en",
                   "year": "-1",
                    "view": type_report
            }
            try:
                if type_report == "bs":
                    new_data = data_loader_balance_sheet(code ,bs, params, "http://cophieu68.vn/incomestatement.php")
                elif type_report == "ist":
                    new_data = data_loader_business_result(code, ist, params, "http://cophieu68.vn/incomestatement.php")
                else:
                    new_data = data_loader_cash_flow(code, cf, params, "http://cophieu68.vn/incomestatement.php")
            except:
                print("---------Caught error with "+ code+ "--" + type_report)
            
            print(cat_index, category, code, type_report)

        write_csv(type_report+'-'+category, new_data, './data/')
```
